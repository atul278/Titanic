# Title

```python
train['Title'] =train['Name'].map(lambda name:name.split(',')[1].split('.')[0].strip())
```

```python
Title_Dictionary = {
                        "Capt":       "Officer",
                        "Col":        "Officer",
                        "Major":      "Officer",
                        "Jonkheer":   "Royalty",
                        "Don":        "Royalty",
                        "Sir" :       "Royalty",
                        "Dr":         "Officer",
                        "Rev":        "Officer",
                        "the Countess":"Royalty",
                        "Dona":       "Royalty",
                        "Mme":        "Mrs",
                        "Mlle":       "Miss",
                        "Ms":         "Mrs",
                        "Mr" :        "Mr",
                        "Mrs" :       "Mrs",
                        "Miss" :      "Miss",
                        "Master" :    "Master",
                        "Lady" :      "Royalty"

                    }
    
```

```python
train["Title"]=train["Title"].map(Title_Dictionary)
```

```python
test['Title'] =test['Name'].map(lambda name:name.split(',')[1].split('.')[0].strip())
test["Title"]=test["Title"].map(Title_Dictionary)
```

# Model 

## ANN

```python
from keras.layers import Dense
from keras.models import Sequential
from keras.layers import Dropout,BatchNormalization,Activation
```

```python
model=Sequential()
```

```python
model.add(Dense(input_dim=27, units=15, kernel_initializer="uniform"))
model.add(BatchNormalization())
model.add(Activation("relu"))
model.add(Dropout(0.4))
model.add(Dense(kernel_initializer="uniform",output_dim=8))
model.add(BatchNormalization())
model.add(Activation("relu"))
model.add(Dropout(0.4))
model.add(Dense(kernel_initializer="uniform",output_dim=4))
model.add(BatchNormalization())
model.add(Activation("relu"))
model.add(Dropout(0.4))
model.add(Dense(kernel_initializer="uniform",output_dim=2))
model.add(BatchNormalization())
model.add(Activation("relu"))
model.add(Dropout(0.4))
model.add(Dense(activation="sigmoid",kernel_initializer="uniform",output_dim=1))
```

```python
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
model.fit(x_train,y_train,batch_size=10,epochs=300)
```
